```python
# A hashlib é uma biblioteca que já vem instalada junto com o python 
# Ela os permite utilizar diversos algoritmos de hashing, sendo o SHA-256 um deles 
# Então, antes de começar o exemplo, vamos importá-la
import hashlib
```

```python
# Aqui dizemos que vamos usar o algoritmo de hash SHA-256
sha256 = hashlib.sha256()

# Para escolhermos a mensagem em cima da qual o hash vai ser feito
# utilizamos a função update
sha256.update(b"This is a sha-256 usage example")
```

```python
# ----------- Curiosidades -----------

# É possível ver o tamanho, em bytes, do hash que será gerado pelo algoritmo
print("Digest size:", sha256.digest_size, "bytes")
# E também o tamanho do bloco interno que o algoritmo de hash utiliza
print("Block size used by the algorithm:", sha256.block_size, "bytes")
```

```python
# Por fim, aqui está o valor do hash aplicado à frase que passamos acima

# Em bytes
print("SHA-256 result as bytes:", sha256.digest())
# Em hexadecimal
print("SHA-256 result in hexa:", sha256.hexdigest())
```
